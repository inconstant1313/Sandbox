*Проанализируем число людей каждого возраста*

```python
print("Уникальные значения в `Age` => {}".format(df.Age.unique()))
```

```python
# Удалим аномальные значения
df = df[(df.Age >= 0) & (df.Age <= 110)]
df.Age.value_counts()
```

```python
plt.figure(figsize=(24,6))
plt.xticks(rotation=90)
ax = sns.countplot(x=df.Age)
ax.set_title("Распределение пациентов по возрасту")
experiment.log_figure(figure=plt)
plt.show()
```

*Выделяем числовые признаки:*

```python
numeric_data = df.copy()
numeric_data.drop([
    'AppointmentID', 
    'PatientId',
    'Gender',
    'ScheduledDay', 
    'AppointmentDay',
    'Neighbourhood', 
    'No-show'
    ], axis=1, inplace = True)
numeric_data.info()
```

*Построим тепловую карту корреляции между числовыми признаками:*

```python
half = np.tril(numeric_data.corr())
ax = sns.heatmap(numeric_data.corr(), annot=True, mask=half, cbar_kws={'label':'Корреляция'},linecolor='lightgrey',linewidths=0.1)
ax.set_title('Тепловая карта корреляции числовых признаков')
ax.set_xlabel('Числовые признаки')
ax.set_ylabel('Числовые признаки')
experiment.log_figure(figure=plt);
```

*Из тепловой матрицы можно сделать вывод, что между числовыми признаками отсутствует мультиколлинеарность. Большая часть из числовых признаков имеют прямую связь, но есть признаки и с обратной связью.*

## Кодировка категориальных признаков

```python
categorical_columns_names = ['Gender', 'Neighbourhood']
```

```python
encoded_columns = pd.get_dummies(df, columns = categorical_columns_names)
```

```python
encoded_columns.drop(['AppointmentID', 'PatientId', 'ScheduledDay', 'AppointmentDay', 'No-show'], axis=1, inplace = True)
```

```python
encoded_columns.head()
```

```python
y = df['No-show']
```

```python
y
```

```python
y = y.replace({'No': 0, 'Yes': 1})
```

```python
y.value_counts(True)
```

## Шкалирование данных

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(encoded_columns)
```

```python
print(X.shape)
```

# Модель (этот раздел будет подробно изучен далее в курсе)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Инициализация модели

```python
logreg = LogisticRegression()
```

```python
random_state = 42
```

### Обучение модели

```python
logreg.fit(X, y)
```

#### Логирование параметров обучения

```python
params={"random_state":random_state,
        "model_type":"logreg"
}
```

```python
experiment.log_parameters(params)
```

### Предсказания модели

```python
y_pred = logreg.predict(X)
```

### Анализ модели

```python
accuracy = accuracy_score(y, y_pred)

print("Доля правильных ответов: {:6.3f}".format(accuracy))
```

# Логирование метрик

```python
metrics = {"accuracy":accuracy}
```

```python
experiment.log_metrics(metrics)
```
